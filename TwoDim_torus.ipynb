```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
#w1, w2 = np.random.uniform(0, 5, 2)
w1, w2 = 2, 3
alpha = np.pi
beta1, beta2 = np.random.uniform(0, 2*np.pi, 2)
```

```python
print(f'{w1=}, {w2=}')
print(f'{beta1=}, {beta2=}')
```

```python
s = np.arange(0, 1000)
phi1 = np.ones_like(s)*alpha
phi2 = beta2 + (w2/w1)*(alpha - beta1 + 2*np.pi*s)
phi2 = phi2 % (2*np.pi)
```

```python

plt.scatter(phi1, phi2, s=0.5, alpha=0.5)

plt.xlim(0, 2*np.pi)
plt.ylim(0, 2*np.pi)
plt.xlabel(r'$\phi_1$')
plt.ylabel(r'$\phi_2$')
plt.title(r'$\alpha = \pi, {}$'+f's={len(s)}')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(0, 2*np.pi)
ax.set_xlabel(r'$\phi_1$')
ax.set_ylabel(r'$\phi_2$')
ax.set_aspect('equal', adjustable='datalim')
points, = ax.plot([], [], 'bo', markersize=0.5, alpha=0.5)

# フレーム更新関数
def update(s):
    points.set_data(phi1[: s], phi2[: s])
    ax.set_title(r'$\alpha = \pi, $'+f'{s=}')
    return points,

# アニメーション作成
ani = FuncAnimation(
    fig,
    update,
    frames=len(s),
    interval=20,  # アニメーションの速度（ms）
    blit=True
)

plt.show()
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=30, metadata=dict(artist='Me'), bitrate=1800)
ani.save('hoge.mp4', writer=writer)

plt.show()
```
